# Работа с папками, файлами
- Кейс: считать файлы из каждой директории и объединить данные в датафрейм, вставив столбцы, которые обозначают название директории и файла

```python
path = Path(a)
df = pd.DataFrame()
path = Path(a)/'data'

#read data from all files
for current_path, dirs, files in os.walk(path):
    for file in files:
        part_path = Path(current_path)
        data_path = part_path / file

        if file != '.DS_Store':
            temp_df = pd.read_csv(data_path)
            
            #extract date and name
            date = data_path.parts[-3]
            name = data_path.parts[-2]
            
            #add date and name columns
            temp_df['name'] = name
            temp_df['date'] = date

            #insert data into main df
            df = pd.concat((df, temp_df))
```

```python
# / - убирает из директории название файла
print(part_path / file)
print(part_path)
```
